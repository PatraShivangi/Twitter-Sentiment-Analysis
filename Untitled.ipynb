```python
df = pd.read_csv('C:/Users/Lenovo/Desktop/task4/twitter_training.csv', header=None)
df.head()
```

```python
df.columns = ["ID", "Topic", "Sentiment", "Text"]
df.head()
```

```python
df = df[["Text", "Sentiment"]]
df.head()
```

```python
import re
import nltk
from nltk.corpus import stopwords

nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

# Text Cleaning Function
def clean_text(text):
    text = str(text).lower()  # Lowercase
    text = re.sub(r'[^a-zA-Z\s]', '', text)  # Special Characters Remove
    text = ' '.join([word for word in text.split() if word not in stop_words])  # Stopwords Remove
    return text

# Apply Cleaning
df["Cleaned_Text"] = df["Text"].apply(clean_text)
df.head()
```

```python
from nltk.sentiment import SentimentIntensityAnalyzer

nltk.download('vader_lexicon')
sia = SentimentIntensityAnalyzer()

# Sentiment Score Assign Karna
df["Sentiment_Score"] = df["Cleaned_Text"].apply(lambda x: sia.polarity_scores(x)["compound"])
df.head()
```

```python
def classify_sentiment(score):
    if score >= 0.05:
        return "Positive"
    elif score <= -0.05:
        return "Negative"
    else:
        return "Neutral"

df["Predicted_Sentiment"] = df["Sentiment_Score"].apply(classify_sentiment)
df.head()
```

```python
from textblob import TextBlob

# TextBlob Sentiment Score Generate
df["TextBlob_Score"] = df["Cleaned_Text"].apply(lambda x: TextBlob(x).sentiment.polarity)
df["TextBlob_Sentiment"] = df["TextBlob_Score"].apply(lambda x: "Positive" if x > 0 else ("Negative" if x < 0 else "Neutral"))
df.head()
```

```python
import matplotlib.pyplot as plt

# Sentiment Count
sentiment_counts = df["TextBlob_Sentiment"].value_counts()

# Bar Chart Plot
plt.figure(figsize=(6, 4))
sentiment_counts.plot(kind="bar", color=["green", "red", "blue"])
plt.xlabel("Sentiment")
plt.ylabel("Count")
plt.title("Sentiment Distribution")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Apply Seaborn Styling
sns.set_theme(style="whitegrid")

# Sentiment Count
sentiment_counts = df["TextBlob_Sentiment"].value_counts()

# Bar Chart Plot
plt.figure(figsize=(8, 5))
sns.barplot(x=sentiment_counts.index, y=sentiment_counts.values, palette="coolwarm")

# Labels & Title
plt.xlabel("Sentiment", fontsize=12, fontweight="bold")
plt.ylabel("Count", fontsize=12, fontweight="bold")
plt.title("Sentiment Distribution (Stylized)", fontsize=14, fontweight="bold")

# Display Chart
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

#Seaborn Styling
sns.set_theme(style="whitegrid")

# Sentiment Count
sentiment_counts = df["TextBlob_Sentiment"].value_counts().reset_index()
sentiment_counts.columns = ["Sentiment", "Count"]  # Column Rename for Clarity

#  Bar Chart Plot
plt.figure(figsize=(8, 5))
sns.barplot(x="Sentiment", y="Count", hue="Sentiment", data=sentiment_counts, palette="coolwarm", legend=False)

# Labels & Title
plt.xlabel("Sentiment", fontsize=12, fontweight="bold")
plt.ylabel("Count", fontsize=12, fontweight="bold")
plt.title("Sentiment Distribution (Stylized & Fixed)", fontsize=14, fontweight="bold")

# Display Chart
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="darkgrid")  # Dark background for contrast

sentiment_counts = df["TextBlob_Sentiment"].value_counts().reset_index()
sentiment_counts.columns = ["Sentiment", "Count"]

fig, ax = plt.subplots(figsize=(8, 5))

bars = sns.barplot(x="Sentiment", y="Count", hue="Sentiment", data=sentiment_counts, palette="coolwarm", legend=False, ax=ax)

for bar in bars.patches:
    bar.set_edgecolor("black")  # Dark edge for contrast
    bar.set_linewidth(1)  # Border thickness
    bar.set_alpha(0.8)  # Slight transparency for modern effect

ax.set_xlabel("Sentiment", fontsize=13, fontweight="bold", color="white")
ax.set_ylabel("Count", fontsize=13, fontweight="bold", color="white")
ax.set_title("Sentiment Distribution (Enhanced)", fontsize=15, fontweight="bold", color="white")

plt.show()
```

```python
import numpy as np
import pandas as pd

# Random Date-Time Generate
np.random.seed(42)  # Seed for reproducibility
df["Timestamp"] = pd.date_range(start="2024-01-01", periods=len(df), freq="D")  # Daily frequency

print(df.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="whitegrid")

plt.figure(figsize=(10, 5))
sns.lineplot(data=df, x="Timestamp", y="TextBlob_Score", hue="TextBlob_Sentiment", marker="o", palette="coolwarm")

plt.xlabel("Date", fontsize=12, fontweight="bold")
plt.ylabel("Sentiment Score", fontsize=12, fontweight="bold")
plt.title("Sentiment Trends Over Time", fontsize=14, fontweight="bold")

plt.xticks(rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="white")
df_daily = df.groupby("Timestamp")["TextBlob_Score"].mean().reset_index()
plt.figure(figsize=(10, 5))
sns.lineplot(data=df_daily, x="Timestamp", y="TextBlob_Score", color="purple", linewidth=2.5, marker="o", markersize=7)

# Horizontal Reference Line for Neutral Sentiment
plt.axhline(0, linestyle="dashed", color="gray", linewidth=1.2)
plt.xlabel("Date", fontsize=13, fontweight="bold")
plt.ylabel("Average Sentiment Score", fontsize=13, fontweight="bold")
plt.title("Sentiment Trends Over Time", fontsize=15, fontweight="bold")

# Rotated X-axis Labels for Better Readability
plt.xticks(rotation=45)
plt.show()
```
